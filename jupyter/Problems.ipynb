```python
#2) area of flower
area= np.cumsum(1/2 * r**2)*(theta[1]-theta[0])
print(area[len(area)-1])
#or
area1= np.sum(1/2 * r**2)*(theta[1]-theta[0])
print(area1)
```

## PROBLEM 10

```python
# to eleminate the non numeric row
a1 = np.array([[10.5, 22.5, 3.8],
                  [41, np.nan, np.nan]])
  
print("Given array:")
print(a1)
  
print("\nRemove all rows containing non-numeric elements")
print(a1[~np.isnan(a1).any(axis=1)])
```

## PROBLEM 12

```python
# Find the most frequent value in a NumPy array
a= np.array([2,3,4,1,6,2,5,2,2,5,6,2])
print(np.bincount(a).max())
# frequency of 4,5,6
print(np.sum(a==4))
print(np.sum(a==5))
print(np.sum(a==6))
print(repr(a).count('2'))
```
