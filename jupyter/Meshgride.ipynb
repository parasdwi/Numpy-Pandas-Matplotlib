```python
# MESHGRIDE IS USE FOR MAPING
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x= np.linspace(0,100,1000)
y= np.linspace(0,100,1000)
```

```python
xv,yv= np.meshgrid(x,y)
z= np.sqrt(xv**2+yv**2) 
```

```python
plt.contourf(xv,yv,z, levels=100)
plt.colorbar()
```
